```python
new_df = df.drop(['RSHA','SGR', 'ROP', 'DTS', 'DCAL', 'MUDWEIGHT', 'RMIC', 'ROPA', 'RXO'], axis=1)
new_df.head()
```

```python
dfna = new_df.dropna()
```

```python
dfna
```

```python
final_df = dfna.drop(['WELL','DEPTH_MD','X_LOC','Y_LOC','Z_LOC','GROUP','FORMATION','FORCE_2020_LITHOFACIES_CONFIDENCE'], axis=1)
```

```python
final_df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,15))
cor= final_df.corr()
ax = sns.heatmap(cor, annot=True, cmap = plt.cm.CMRmap_r)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.show()
```
